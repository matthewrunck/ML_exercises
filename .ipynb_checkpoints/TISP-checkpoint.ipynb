### Thresholding-based Iterative Selection Procedure
Introduces a thresholdiing function to select features: $$ \Theta(x,\lambda) = \begin{cases} 0 & |x| < \lambda\\ x & |x|\geq \lambda \end{cases}$$

```python
def theta(g,lambda1):
    thet = np.zeros(np.shape(g))
    thet[np.absolute(g) <= lambda1]=0
    thet[np.absolute(g) > lambda1]= g[np.absolute(g) > lambda1]
    return thet
```

We still use gradient ascent to maximise the likelihood-
$$w^{(temp)}=w^{(t)}+\eta X^T[y-\frac{1}{1+exp(-Xw^{(t)})}]$$

```python
def dL(w,X,y):
    delt = np.dot((y-1/(1+np.exp(-np.dot(X,w)))).T,X)
    return delt.T
```

but now we use the thresholding function to retain only those featres with weights greater than $\lambda$
$$w^{(t+1)}=\Theta(w^{(temp)},\lambda)$$


```python
def TISP(train,train_label,test,test_label,lambda1,iterations=100,eta=1):
    #Initiliase weights
    w =np.zeros((test.shape[1],1))

    # Set parameters
    
    N = len(train_label)
    features = np.zeros(len(lambda1))
    train_pred = np.zeros((N,len(lambda1)))
    test_pred = np.zeros((len(test_label),len(lambda1)))
    for j in np.arange(0,len(lambda1)):

        for i in np.arange(0,iterations): 
            #Temp weights by gradient ascent
            v =w+ eta*dL(w,train,train_label)/len(train_label)
            #Apply thresholding function
            w = theta(v,lambda1[j]) 



        train_pred[:,j] = (np.dot(train,w)>0)[:,0]
        test_pred[:,j] = (np.dot(test,w)>0)[:,0]
        features[j] = np.size(w[w!=0])

    tl =np.tile(train_label,(len(lambda1)))
    train_error = np.sum(train_pred!=tl,axis=0)/len(train_label)

    vl =np.tile(test_label,(len(lambda1)))
    test_error = np.sum(test_pred!=vl,axis=0)/len(test_label)

    plt.plot(features,train_error,'r-',label='training set',)
    plt.plot(features,test_error,'b--',label='test set')
    plt.title('Features vs error')
    plt.xlabel('No. of features selected')
    plt.ylabel('Error')
    leg = plt.legend(loc='best', ncol=2)
    leg.get_frame()
    return train_pred, test_pred,train_error,test_error
```

```python
#load training and test data
gis_train,gis_train_labels,gis_test,gis_test_labels =load_data("Data\\gisette_train.data","Data\\gisette_train.labels","Data\\gisette_valid.data","Data\\gisette_valid.labels")
mad_train,mad_train_labels,mad_test,mad_test_labels = load_data("Data\\madelon_train.data","Data\\madelon_train.labels","Data\\madelon_valid.data","Data\\madelon_valid.labels")
```

```python
gis_train_pred,gis_test_pred,gis_train_error,gis_test_error =TISP(gis_train,gis_train_labels,gis_test,gis_test_labels,lambda1,iterations=100,eta=1)
```
