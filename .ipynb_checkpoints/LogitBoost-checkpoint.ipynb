### LogitBoost
'Boosting' is a means of combining weak classifiers into a single strong classifier.
Essentially at each step, we choose a weak classifier on a single feature that gives the best reduction in the loss function to include.
Then the new predictions our calculated from the updated classifier and the samples are rewiegihted so that misclassified observations received more weight, so that subseuent classifiers are chosen that do better at pediciting the currently misclassified observations.
At each iteration, we us our current estimates of the probability of each observation $p = \frac{1}{1+exp(-X\beta_{all})}$ and perform simple linear regression on the reweighted samples $z=\frac{y-p}{p(1-p)}$ for each of the M features to find the feature $X_j$ which minimizes the logistic loss function if $\beta_j$ is included in the update to $\beta_{all}$ 

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
from Load_data import load_data
```

```python
#Load and clean  training data
train_data = pd.read_csv("Data\\gisette_train.data", delimiter=' ',header=None)
train = train_data.dropna(axis='columns')

train_std= np.std(train,axis=0)

col_to_drop = train_std==0

train_label = np.array(pd.read_csv("Data\\gisette_train.labels", delimiter=' ',header=None))


train = train.drop(train.columns[col_to_drop],axis=1)

train = np.c_[np.ones(train.shape[0]),train]
train = np.array(train)
N = train.shape[0]
M = train.shape[1]
#Load and clean testing data


valid_data = pd.read_csv("Data\\gisette_valid.data", delimiter=' ',header=None)
valid_label = pd.read_csv("Data\\gisette_valid.labels", delimiter=' ',header=None)
valid = valid_data.dropna(axis='columns')


valid = valid.drop(valid.columns[col_to_drop],axis=1)

valid = np.c_[np.ones(valid.shape[0]),valid]
```

```python
#Initialising variables
beta_all =np.zeros((train.shape[1],1))
Loss = np.zeros(300)
train_error = np.zeros(4)
test_error = np.zeros(4)
ik = 0
```

```python

for i in np.arange(0,300):
    H = np.dot(train,beta_all) 
    p=1/(1+np.exp(-2*H)) #Current estimates 
    w = p*(1-p) #weights for next estimation
    z = (0.5*(train_label+1)-p)/w #reweighted samples
    beta_j = np.zeros((2,M-1)) #coefficients for linear classifiers for each feature
    
    #Calculating the weighted least squares estimate for each classifier manually
    a = np.sum(w)
    b = np.sum(w*train[:,1:],axis=0)
    c = np.sum(w*(train[:,1:]**2),axis=0)
    d = np.sum(w*z)
    e = np.sum(w*train[:,1:]*z,axis=0)

    l = a*c-b**2 
    beta_j[0,l!=0] = ((c*d-b*e)/l)[l!=0]
    beta_j[1,l!=0]= ((a*e-b*d)/l)[l!=0]
    beta_j[0,l==0] = (d/a)
    beta_j[1,l==0] = 0
    
    #Updated  loss if each feature is included
    Hj = H + 0.5*(beta_j[0]+beta_j[1]*train[:,1:])
    loss_j =np.sum(np.log(1+np.exp(-2*train_label*Hj)),0)
    
    coef = beta_j
    newloss = loss_j
    #find the classifier with smalest loss
    J = np.argmin(newloss)
    #update classifier 
    beta_all[0] = beta_all[0] + 0.5*coef[0,J]
    beta_all[J+1] = beta_all[J+1] + 0.5*coef[1,J]
    #Loss for updated classifier
    Loss[i] = newloss[J]
    
    #get results for select iterations
    if i+1 in [10, 30, 100, 300]:
        train_predict = np.sign(np.dot(train,beta_all))
        train_error[ik] = np.mean((train_predict!= train_label))
        valid_predict = np.sign(np.dot(valid,beta_all))
        test_error[ik] = np.mean((valid_predict!= valid_label))
        ik=ik+1
```

```python
plt.plot(np.arange(0,300),Loss,'b-')
plt.title('Loss function vs iterations')
plt.xlabel('No. of iterations')
plt.ylabel('Loss function')
```
