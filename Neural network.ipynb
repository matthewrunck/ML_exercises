#### Neural Network
Implementing a simple Neural net in the MiniBooNE dataset from Fermilab.
This is a NN with two hidden layers with 128 neurons in each layer, using the adam optimizer with 300 epoch and batches of size 4.
The model is trained on 10 random 80-20 train-test splits of the original data, with the final accuracy being the average performace across all 10 splits.

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Load data
train_data = np.array(pd.read_csv("Data\\MiniBooNE_PID.txt", delim_whitespace=True,header=None,skiprows=1))
train_mean = np.mean(train_data,axis=0)

train_std= np.std(train_data,axis=0)


train = (train_data-train_mean)/train_std


train = np.c_[np.ones(train.shape[0]),train]

#In the dataset the first 36499 obs are from electron neutrinos (signal) the rest are muon neutrinos (noise)
labels=np.concatenate((np.ones((36499,1)),np.zeros((93565,1))),axis=0)
```

```python
#Use 80% of the data as training set
c = int(0.8*train.shape[0])

#initializing vectors to hold the accuracy for each split
testacc = np.zeros(10)
trainacc = np.zeros(10)
```

```python
for i in np.arange(0,10):
    model = tf.keras.Sequential()
    model.add(layers.Dense(128, activation='relu')) #add hidden layer with 'relu' activation function
    model.add(layers.Dense(128, activation='relu')) #add second hidden layer
    model.add(layers.Dense(1)) #output layer
    
    #randomly permute sample date
    indices = np.random.permutation(train.shape[0])
    #take first 80% of random observations as training set, the rest as test
    training_idx, test_idx = indices[:c], indices[c:]
    training, test = train[training_idx,:], train[test_idx,:]
    train_labels, test_labels = labels[training_idx,:], labels[test_idx,:]
    
    #compile and train the model
    model.compile(optimizer=tf.train.AdamOptimizer(0.001),loss='binary_crossentropy',metrics=['accuracy'])
    model.fit(training, train_labels, epochs=100, batch_size=1024,verbose=0)
    
    #record accuracy for current split
    trainacc[i] = model.evaluate(training,train_labels)[1]
    testacc[i] = model.evaluate(test,test_labels)[1]
    print(i)

a = 1-np.mean(trainacc)
b = 1-np.mean(testacc)
```
